```python
def main(rcpno: str) -> None:
    url = f"https://dart.fss.or.kr/dsaf001/main.do?rcpNo={rcpno}"
    req = requests.get(
        url,
        "lxml",
    ).text
    dcmNo = find_dcmNo(req)
    max_id = find_max_eleid(req)
    res = ""
    for id in range(max_id):
        id += 1
        src_url = f"https://dart.fss.or.kr/report/viewer.do?rcpNo={rcpno}&dcmNo={dcmNo}&eleId={id}&offset=800&length=4053&dtd=dart3.xsd"
        text = text_from_url(src_url)
        text = del_blank(text)
        res += text + " "

    with open(f"{rcpno}.txt", "w") as f:
        f.write(res)
```
